```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_line_chart(exec_time, title, fig_name):
    plt.subplots(figsize=(12, 6))
    plt.plot([100, 1000, 10000, 100000], exec_time[0], ':', linewidth=2, label='BubbleSort')
    plt.plot([100, 1000, 10000, 100000], exec_time[1], ':', linewidth=2, label='SelectionSort')
    plt.plot([100, 1000, 10000, 100000], exec_time[2], ':', linewidth=2, label='InsertionSort')
    plt.plot([100, 1000, 10000, 100000], exec_time[3], linewidth=2, label='CocktailSort')
    plt.title(title, fontsize=14)
    plt.xlabel('Tamanho do Vetor', fontsize=14)
    plt.ylabel('Tempo de execução (ms)', fontsize=14)
    plt.tick_params(axis='both', which='major', labelsize=12)
    plt.yscale('log')
    plt.xscale('log')
    plt.legend()
    plt.grid(True)
    
    fig2 = plt.gcf()
    fig2.savefig(fig_name, format='png')
    
    plt.show()
```

```python
exec_time = [
    [0.2070, 6.543667, 614.108, 61273.7058],
    [0.0835, 5.028667, 307.2975, 30785.2654],
    [0.0045, 0.009667, 0.0955, 0.9972],
    [0.005, 0.009667, 0.10225, 0.9128]
]
plot_line_chart(exec_time, 'Vetor Ordenado', 'vetor_ordenado.png')
```
