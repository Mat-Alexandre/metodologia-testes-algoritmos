```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_line_chart(exec_time, title, fig_name):
    fig, ax =plt.subplots(figsize=(12, 6))
    x = [100, 1000, 10000, 100000]
    ax.plot(x, exec_time[0], ':', linewidth=4, label='BubbleSort')
    ax.plot(x, exec_time[1], ':', linewidth=4, label='SelectionSort')
    ax.plot(x, exec_time[2], ':', linewidth=4, label='InsertionSort')
    ax.plot(x, exec_time[3], linewidth=4, label='CocktailSort')
    
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Tamanho do Vetor', fontsize=14)
    ax.set_ylabel('Ordem de grandeza (ms)', fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.legend()
    ax.grid(True)
    
    fig = plt.gcf()
    fig.savefig('../DADOS/'+fig_name, format='png')
    
    plt.show()
```

```python
exec_time = [
    [0.1562, 6.4324, 631.7196, 62581.951],
    [0.0836, 3.9638, 312.6686, 30941.3608],
    [0.0032, 0.0094, 0.0918, 0.961],
    [0.0022, 0.0094, 0.0902, 0.9436]
]
plot_line_chart(exec_time, 'Vetor Ordenado', 'vetor_ordenado.png')
```
